## 1. Setup and Configuration <a id='setup'></a>

````python
import os
import json
from openai import OpenAI
from typing import Optional, Dict, Any, List
import time

# Initialize OpenAI client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# Helper function to create and poll responses
def create_gpt5_response(
    messages: List[Dict[str, str]],
    reasoning_effort: str = "medium",
    tools: Optional[List[Dict]] = None,
    verbosity: Optional[str] = None,
    previous_response_id: Optional[str] = None
) -> Dict[str, Any]:
    """Create and poll a GPT-5 response using the Responses API."""
    
    # Prepare request parameters
    params = {
        "model": "gpt-5.5",
        "input": messages,
        "reasoning": {"effort": reasoning_effort}
    }
    
    if tools:
        params["tools"] = tools
    if verbosity:
        params["text"] = {"verbosity": verbosity}
    if previous_response_id:
        params["previous_response_id"] = previous_response_id
    
    # Create response
    response = client.responses.create(**params)
    
    # Poll for completion
    while response.status != "completed":
        time.sleep(1)
        response = client.responses.retrieve(response.id)
    
    return response

print("Setup complete. Ready to demonstrate GPT-5 prompting techniques.")
````

## 2. Controlling Agentic Eagerness <a id='eagerness'></a>

GPT-5 can be calibrated for different levels of autonomy and proactivity.

### Example: Less Eagerness (Quick, Focused Response)

````python
# Less eager prompt - for quick, focused responses
less_eager_system = """
<context_gathering>
Goal: Get enough context fast. Parallelize discovery and stop as soon as you can act.

Method:
- Start broad, then fan out to focused subqueries.
- In parallel, launch varied queries; read top hits per query. Deduplicate paths and cache; don't repeat queries.
- Avoid over searching for context. If needed, run targeted searches in one parallel batch.

Early stop criteria:
- You can name exact content to change.
- Top hits converge (~70%) on one area/path.

Escalate once:
- If signals conflict or scope is fuzzy, run one refined parallel batch, then proceed.

Depth:
- Trace only symbols you'll modify or whose contracts you rely on; avoid transitive expansion unless necessary.

Loop:
- Batch search → minimal plan → complete task.
- Search again only if validation fails or new unknowns appear. Prefer acting over more searching.
</context_gathering>
"""

# Example tools for demonstration (Responses API flat format — no "function" wrapper)
search_tools = [
    {
        "type": "function",
        "name": "search_codebase",
        "description": "Search for code patterns in the codebase",
        "parameters": {
            "type": "object",
            "properties": {
                "query": {"type": "string", "description": "Search query"},
                "path": {"type": "string", "description": "Optional path to search in"}
            },
            "required": ["query"]
        }
    },
    {
        "type": "function",
        "name": "read_file",
        "description": "Read a file from the codebase",
        "parameters": {
            "type": "object",
            "properties": {
                "path": {"type": "string", "description": "File path to read"}
            },
            "required": ["path"]
        }
    }
]

response = create_gpt5_response(
    messages=[
        {"role": "system", "content": less_eager_system},
        {"role": "user", "content": "Find and fix the authentication bug in the login module"}
    ],
    reasoning_effort="low",  # Lower reasoning for faster response
    tools=search_tools
)

print("Less Eager Response (focused, quick):")
print(str(response.output)[:1000] + "...")
````

### Example: More Eagerness (Autonomous, Persistent)

````python
# More eager prompt - for autonomous, persistent behavior
more_eager_system = """
<persistence>
- You are an agent - please keep going until the user's query is completely resolved, before ending your turn and yielding back to the user.
- Only terminate your turn when you are sure that the problem is solved.
- Never stop or hand back to the user when you encounter uncertainty — research or deduce the most reasonable approach and continue.
- Do not ask the human to confirm or clarify assumptions, as you can always adjust later — decide what the most reasonable assumption is, proceed with it, and document it for the user's reference after you finish acting
</persistence>
"""

response = create_gpt5_response(
    messages=[
        {"role": "system", "content": more_eager_system},
        {"role": "user", "content": "Refactor the authentication system to use OAuth 2.0"}
    ],
    reasoning_effort="high",  # Higher reasoning for thorough exploration
    tools=search_tools
)

print("More Eager Response (autonomous, persistent):")
print(str(response.output)[:1000] + "...")
````

## 3. Tool Preambles <a id='preambles'></a>

GPT-5 is trained to provide clear upfront plans and progress updates via tool preambles.

````python
# Tool preambles configuration
tool_preamble_system = """
<tool_preambles>
- Always begin by rephrasing the user's goal in a friendly, clear, and concise manner, before calling any tools.
- Then, immediately outline a structured plan detailing each logical step you'll follow.
- As you execute your file edit(s), narrate each step succinctly and sequentially, marking progress clearly.
- Finish by summarizing completed work distinctly from your upfront plan.
</tool_preambles>
"""

# File editing tools for demonstration (Responses API flat format)
file_tools = [
    {
        "type": "function",
        "name": "edit_file",
        "description": "Edit a file in the codebase",
        "parameters": {
            "type": "object",
            "properties": {
                "path": {"type": "string"},
                "content": {"type": "string"}
            },
            "required": ["path", "content"]
        }
    }
]

response = create_gpt5_response(
    messages=[
        {"role": "system", "content": tool_preamble_system},
        {"role": "user", "content": "Add input validation to the user registration form"}
    ],
    tools=file_tools
)

print("Response with Tool Preambles:")
for output in response.output[:3]:  # Show first few outputs
    if output.type == "message":
        text = output.content[0].text if output.content else ""
        print(f"Preamble: {text}")
    elif output.type == "function_call":
        print(f"Tool Call: {output.name} - {output.arguments[:100]}...")
````

## 4. Reasoning Effort <a id='reasoning'></a>

The `reasoning_effort` parameter controls how hard the model thinks.

````python
# Compare different reasoning effort levels
test_message = "Implement a binary search algorithm with proper error handling"

reasoning_levels = ["low", "medium", "high"]
results = {}

for level in reasoning_levels:
    response = create_gpt5_response(
        messages=[
            {"role": "user", "content": test_message}
        ],
        reasoning_effort=level
    )
    results[level] = response
    print(f"\nReasoning Effort: {level}")
    print(f"Reasoning tokens: {getattr(response.usage, 'output_tokens_details', None)}")
    print(f"Output preview: {response.output_text[:200]}...")
````

### Reusing Reasoning Context with previous_response_id

````python
# Initial response with reasoning
initial_response = create_gpt5_response(
    messages=[
        {"role": "user", "content": "Design a REST API for a task management system"}
    ],
    reasoning_effort="high"
)

print(f"Initial Response ID: {initial_response.id}")
print(f"Output tokens used: {initial_response.usage.output_tokens}")

# Follow-up using previous reasoning
followup_response = create_gpt5_response(
    messages=[
        {"role": "user", "content": "Now implement the POST endpoint for creating tasks"}
    ],
    reasoning_effort="medium",
    previous_response_id=initial_response.id  # Reuse previous reasoning
)

print(f"\nFollow-up Response ID: {followup_response.id}")
print(f"Output tokens used: {followup_response.usage.output_tokens}")
print("Note: Reasoning context from previous response is reused, reducing token usage")
````

Codex / `gpt-5.3-codex` guidance favors: **persistence** (finish the task before yielding), **verify-before-finish** (run build/tests, read the diff), **brief preambles** (1-2 sentences only at major phase changes), and **patch-based edits** (diffs, not full rewrites).

````python
# Codex-style agentic coding prompt (patterns from the Codex prompting guide)
codex_prompt = """
You are a coding agent operating in a real repository.

<persistence>
Keep going until the user's request is COMPLETELY resolved before ending your turn.
Do not hand back control with the task half-done.
</persistence>

<verification>
Before declaring done: run the build and tests, then read the resulting diff.
If anything fails, fix it and re-verify. Never claim success you haven't observed.
</verification>

<preambles>
Post a brief 1-2 sentence update ONLY when starting a new major phase of work,
or when you discover something that changes the plan. Otherwise stay quiet and work.
</preambles>

<edits>
Express file changes as patches/diffs against existing files, not full-file rewrites.
</edits>
"""

resp = client.responses.create(
    model="gpt-5.5",          # use gpt-5.3-codex in a Codex-style environment
    instructions=codex_prompt,
    input="Add input validation to the create_user endpoint and update its tests.",
    reasoning={"effort": "high"},
)
print(resp.output_text)
````

### Prompt Personalities <a id='personalities'></a>

A **prompt personality** is a reusable persona block that fixes tone, default verbosity, and decision style for a whole product surface. Keep a small library and select per use-case.

````python
# A small library of reusable prompt personalities
PERSONALITIES = {
    "terse_senior_engineer": """
        You are a terse senior engineer. Answer in <=4 sentences or <=5 bullets.
        Prefer the direct fix over discussion. Call out the one risk that matters, nothing else.
    """,
    "patient_mentor": """
        You are a patient mentor. Explain the 'why' before the 'how'.
        Use one small analogy, then a concrete example. Encourage, never condescend.
    """,
    "compliance_reviewer": """
        You are a careful compliance reviewer. Flag unstated assumptions and ungrounded claims.
        Default to caution; never use absolute guarantees. End with an explicit Caveats list.
    """,
}

def ask_as(personality: str, prompt: str, effort: str = "low"):
    return client.responses.create(
        model="gpt-5.5",
        instructions=PERSONALITIES[personality],
        input=prompt,
        reasoning={"effort": effort},
    ).output_text

print(ask_as("terse_senior_engineer", "Should we add a Redis cache in front of this query?"))
````

## 11. `output_verbosity_spec` (Structured Verbosity Control) <a id='verbosity-spec'></a>

The original GPT-5 guide gave you one coarse knob: the `verbosity` API parameter (low/medium/high). The GPT-5.2+ guidance adds a **prompt-level verbosity contract** that is far more controllable.

The pattern: state an explicit default shape for answers -- e.g. *"3-6 sentences or <=5 bullets for typical answers"* -- tell the model to break content into **digestible chunks**, and explicitly forbid **long narrative paragraphs**.

````python
# output_verbosity_spec: an explicit, in-prompt verbosity contract
output_verbosity_spec = """
<output_verbosity_spec>
- Default to 3-6 sentences OR <= 5 bullets for typical answers.
- Break content into digestible chunks; one idea per bullet.
- Avoid long narrative paragraphs unless the user explicitly asks for prose.
- Lead with the answer; put caveats after, not before.
</output_verbosity_spec>
"""

resp = client.responses.create(
    model="gpt-5.5",
    instructions=output_verbosity_spec,
    input="Why might a team choose PostgreSQL over MongoDB?",
    reasoning={"effort": "low"},
)
print(resp.output_text)
````

## 12. Scope Discipline <a id='scope-discipline'></a>

GPT-5.5 (like 5.2) tends to *over-produce* -- adding features and embellishments you never asked for. The fix is a dedicated **scope discipline** block: implement **exactly and only** what was requested.

````python
# Scope discipline guardrail: implement EXACTLY and ONLY what is requested
scope_discipline = """
<scope_discipline>
Implement EXACTLY and ONLY what the user requests.
- No extra features. No added components. No UX embellishments.
- Do not anticipate future requirements or "nice to haves".
- If a requirement is ambiguous, ask one clarifying question instead of guessing big.
</scope_discipline>
"""

resp = client.responses.create(
    model="gpt-5.5",
    instructions=scope_discipline,
    input="Write a Python function that returns the nth Fibonacci number.",
    reasoning={"effort": "low"},
)
print(resp.output_text)  # Expect: one function, no CLI, no tests, no extras.
````

## 14. Uncertainty Self-Checks <a id='uncertainty'></a>

For high-risk domains (medical, legal, financial), add an **uncertainty self-check**: prompt the model to flag **unstated assumptions**, **ungrounded numbers/claims**, and **overly strong language** before finalizing.

````python
# Uncertainty self-check for high-risk domains
uncertainty_self_check = """
<uncertainty_self_check>
Before finalizing, review your answer and explicitly flag:
- Unstated assumptions you relied on.
- Any specific number or claim NOT grounded in the provided context.
- Overly strong language ("always", "guaranteed", "never") that should be softened.
List these under a "Caveats" heading. If there are none, write "Caveats: none".
</uncertainty_self_check>
"""

resp = client.responses.create(
    model="gpt-5.5",
    instructions=uncertainty_self_check,
    input="Estimate the monthly cost of running a 3-node Postgres cluster on a major cloud.",
    reasoning={"effort": "medium"},
)
print(resp.output_text)
````

## Summary and Best Practices

This notebook demonstrated the key prompting techniques for GPT-5.5:

1. **Reasoning Effort**: Use `low` for speed, `medium` for balance, `high` for complex tasks
2. **Verbosity**: Control output length with the API parameter and prompt overrides
3. **Tool Preambles**: Request clear progress updates for better UX
4. **Durable Conversations**: Use `client.conversations.create()` — prefer this over `previous_response_id` for multi-turn context
5. **Clear Instructions**: Avoid contradictions and ambiguity in prompts
6. **Self-Reflection**: Use internal rubrics for high-quality outputs
7. **Metaprompting**: Let GPT-5.5 help optimize its own prompts

### Key Takeaways:
- GPT-5.5 is highly steerable but requires clear, non-contradictory instructions
- The Conversations API eliminates manual history rebuilding and improves coherence
- Different reasoning efforts suit different use cases — `low` is the fastest supported level
- Verbosity can be controlled globally and locally for optimal output
- Production systems benefit from careful prompt engineering and testing
